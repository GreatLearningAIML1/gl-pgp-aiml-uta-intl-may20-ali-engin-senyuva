### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint :

1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.

2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.

3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
df1  = user_data.merge(rating_data,on='user id',how='inner') # Create df using merge function on rating data merge with user id
df2 = movie_data.merge(df1,on='movie id',how='inner') # create another df using merge function on movie data merge with pervious df
df2.head() # see the final result only 5 rows
```

Reasoning:
user data and movie data  as well as rating data are merged. Column value reached to 30, which is 4 from user data, 22 from movie data and 5 from rating data makes 31 but movie id is common for both data frame, so minus 1 makes 30 columns. 

```python
df2.info() # information on final merged df
```

```python
movie_all = df2.groupby(['movie id','movie title'])['rating'].count().reset_index()# create new df using groupby function 
#for movie id and movie title group by rating count

movie_top25 = list(movie_all[movie_all['rating']>100]['movie title']) # create a list from df for movies rating more than 100 
# on movie title
movie_top25
```

```python
avg_rating = df2.groupby(['movie id','movie title'])['rating'].mean().sort_values(ascending=False).reset_index()# find means of 
# the ratings in the df created recently and list descending order
avg_rating[avg_rating['movie title'].isin(movie_top25)].head(25)# select top 25 movies from the list using isin function
```
